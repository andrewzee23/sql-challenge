```python
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
from keys import password
engine = create_engine(f'postgresql://postgres:{password}@localhost:5433/employee_hw')
conn = engine.connect()
```

```python
# salary table (histogram of salary ranges)
salary_df = pd.read_sql('select * from salaries', conn)
salaries = list(salary_df['salary'])
plt.hist(salaries, edgecolor = 'black')
plt.title('Salary Histogram')
plt.xlabel('Salaries')
plt.ylabel('# of Employees')
```

```python
#salary and department titles (bar chart avg salary by title)
employee_dets_df = pd.read_sql('select * from emp_til_sal', conn)



#sorting dfs by title
staff_df = employee_dets_df[(employee_dets_df['title'] == 'Staff')]
senior_engineer_df = employee_dets_df[(employee_dets_df['title'] == 'Senior Engineer')]
engineer_df = employee_dets_df[(employee_dets_df['title'] == 'Engineer')]
senior_staff_df = employee_dets_df[(employee_dets_df['title'] == 'Senior Staff')]
technique_leader_df = employee_dets_df[(employee_dets_df['title'] == 'Technique Leader')]
assistant_engineer_df = employee_dets_df[(employee_dets_df['title'] == 'Assistant Engineer')]
manager_df = employee_dets_df[(employee_dets_df['title'] == 'Manager')]


#finding average salaries for each title
staff_avg = round(staff_df['salary'].sum() / staff_df['salary'].count(),2)
senior_engineer_avg = round(senior_engineer_df['salary'].sum() / senior_engineer_df['salary'].count(),2)
engineer_avg = round(engineer_df['salary'].sum() / engineer_df['salary'].count(),2)
senior_staff_avg = round(senior_staff_df['salary'].sum() / senior_staff_df['salary'].count(),2)
technique_leader_avg = round(technique_leader_df['salary'].sum() / technique_leader_df['salary'].count(),2)
assistant_engineer_avg = round(assistant_engineer_df['salary'].sum() / assistant_engineer_df['salary'].count(),2)
manager_avg = round(manager_df['salary'].sum() / manager_df['salary'].count(),2)


#plotting
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
label = ['Engineer','Senior Engineer','Technique Leader','Staff','Senior Staff','Assistant Engineer','Manager']
totals = [engineer_avg, senior_engineer_avg, technique_leader_avg, staff_avg, senior_staff_avg, assistant_engineer_avg, manager_avg]
colors = ['green', 'yellow', 'orange', 'lightskyblue', 'red', 'teal', 'purple']
plt.title('Avg Salary by Title')
plt.xlabel('Titles')
plt.ylabel('Avg Salary')
ax.bar(label,totals, color = colors)
plt.xticks(label, rotation='vertical')
plt.show()
```
